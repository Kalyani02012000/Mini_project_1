```python
import pandas as pd
# Clean the data

def convert_voting(voting_str):
    voting_str = voting_str.replace('(', '').replace(')', '').strip()
    if 'K' in voting_str:
        return int(float(voting_str.replace('K', '').strip()) * 1000)
    elif 'M' in voting_str:
        return int(float(voting_str.replace('M', '').strip()) * 1_000_000)
    else:
        return int(voting_str.strip())

    

    
def convert_duration(duration_str):
    hours = 0
    minutes = 0
    if 'h' in duration_str:
        hours = int(duration_str.split('h')[0].strip())
        duration_str = duration_str.split('h')[1]
    if 'm' in duration_str:
        minutes = int(duration_str.split('m')[0].strip())
    return hours * 60 + minutes

#read csv_file
df = pd.read_csv("Animation_df.csv")

cleaned_data = []

for index, row in df.iterrows():
    try:
        title = row['Title']
        rating = float(row['Rating'])
        voting = convert_voting(row['Voting'])  
        duration = convert_duration(row['Duration'])
        genre = row['Genre']
        
        cleaned_data.append((title, rating, voting, duration, genre))
    except Exception as e:
        print(f"Skipping row {index} due to error: {e}")

mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="1q2w3e",
  database="imdb"
)
print(mydb)

mycursor = mydb.cursor()

insert_query = """INSERT INTO animation_movies(title, rating, voting, duration, genre)
 VALUES (%s, %s,%s,%s,%s)"""
mycursor.executemany(insert_query,cleaned_data)

mydb.commit()

print(mycursor.rowcount, "record inserted.")
```

```python
import pandas as pd

# Convert functions
def convert_voting(voting_str):
    if isinstance(voting_str, str):
        voting_str = voting_str.replace('(', '').replace(')', '').strip()
        if 'K' in voting_str:
            return int(float(voting_str.replace('K', '').strip()) * 1000)
        elif 'M' in voting_str:
            return int(float(voting_str.replace('M', '').strip()) * 1_000_000)
        else:
            return int(voting_str.strip())
    return 0

def convert_duration(duration_str):
    if not isinstance(duration_str, str):
        return 0
    hours = 0
    minutes = 0
    if 'h' in duration_str:
        parts = duration_str.split('h')
        hours = int(parts[0].strip())
        duration_str = parts[1]
    if 'm' in duration_str:
        minutes = int(duration_str.split('m')[0].strip())
    return hours * 60 + minutes

# File list
file_list = ['Animation_df.csv', 'Sci-Fi_df.csv', 'Adventure_df.csv', 'Fantasy_df.csv', 'Family_df.csv']

# Read and clean all
merged_data = []
for file in file_list:
    df = pd.read_csv(file)
    for index, row in df.iterrows():
        try:
            title = row['Title']
            rating = float(row['Rating'])
            voting = convert_voting(row['Voting'])
            duration = convert_duration(row['Duration'])
            genre = row['Genre']
            merged_data.append((title, rating, voting, duration, genre))
        except Exception as e:
            print(f"Skipping row {index} in {file} due to: {e}")

mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="1q2w3e",
  database="imdb"
)
print(mydb)

mycursor = mydb.cursor()

insert_query = """
INSERT INTO all_movies (title, rating, voting, duration, genre)
VALUES (%s, %s, %s, %s, %s)
"""
mycursor.executemany(insert_query, merged_data)
mydb.commit()

print(f"{mycursor.rowcount} records inserted into 'all_movies'.")
```
